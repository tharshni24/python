### ATM Withdrawal Simulation

Write a Python program that:

- Takes account balance and withdrawal amount as input.
- If the withdrawal amount is more than the balance, print "Insufficient Funds".
- If the withdrawal amount is a multiple of ₹100, deduct it from the balance and print the updated balance.
- Otherwise, print "Enter an amount in multiples of ₹100".

```python
balance = int(input("enter the balance amount in your bank : "))
cash=int(input("enter the amount to be withdrawl from your account : "))
if(cash>balance) :
    print("Insufficient Funds in your account ")
elif (cash%100==0) :
    balance = balance - cash
    print("the balance amount in your account is : ",balance)
else :
    print("Enter an amount in multiples of 100 ")
```

### Leap Year
The rules for determining whether or not a year is a leap year follow: 

- Any year that is divisible by 400 is a leap year. 
- Of the remaining years, any year that is divisible by 100 is not a leap year. 
- Of the remaining years, any year that is divisible by 4 is a leap year. 
- All other years are not leap years. 



Write a program that reads a year from the user and displays a message indicating whether or not it is a leap year. 

Sample Input: 1 <br>
1900  <br>
Sample Output 1  <br>
1900 is not a leap year.  <br>
Sample Input 2  <br>
2000  <br>
Sample Output 2  <br>
2000 is a leap year. 

```python
year = int (input("enter the year :"))
if(year%4==0):
    if(year%100==0):
        if(year%400==0):
            print(year," is a Leap Year")
        else :
            print(year," is not a Leap Year")
    else :
        print(year," is a Leap Year")
else :
    print(year," is not a Leap Year")
```

### Triangle Validity Check

 To make sure the angles form a proper triangle, you need to check two things:

- Each angle should be a positive number (greater than 0).
- The sum of all three angles must be exactly 180 degrees.

Write a Python program to check if the angles you enter can form a valid triangle.

Test Case:

Input:

Angle1 = 60 <br>
Angle2 = 70 <br>
Angle3 = 50<br>

Expected Output: 

The angles form a triangle.

Explanation:

- The program asks you to enter three angles.
- It adds the three angles together.
- Then, it checks if the total is exactly 180 degrees.
- Also, it checks if each angle is greater than 0 (positive).
- If both conditions are true, it tells you that the angles can form a triangle.
- If either condition is false, it tells you that the angles cannot form a triangle.

```python
a1=int(input("Enter angle 1 :"))
a2=int(input("Enter angle 2 :"))
a3=int(input("Enter angle 3 :"))
sum = a1+a2+a3
if (a1>0 and a2>0 and a3>0):
    if(sum==180):
        print("The angles form an triangle")
    else :
        print("These angles cannot form a triangle")
else :
    print("These angles cannot form a triangle")
```

### Train Ticket Fare Based on Age and Distance

A railway system needs a ticket price calculator. The fare depends on the distance of travel and the passenger's age.

Conditions:

- If the distance is more than 100 km:
- If age < 12 or age > 60, apply 50% discount on base fare (₹5 per km).
- If the distance is 100 km or less:
- Children (<12) or senior citizens (>60) get 25% discount.Others pay full fare.

Task:

Write a Python program that takes: <br>

Distance to travel <br>

Passenger's age <br>

And calculates and prints the ticket price. <br>

Sample Input:

Enter distance to travel (km): 120  <br>
Enter age: 65

Expected Output: <br>
Ticket price: ₹300.0   <br>

Explanation: <br>
Base fare = 100*120 × 5 = ₹600  <br>
Since age > 60 and distance > 100 km → 50% discount → ₹300

```python
distance = int(input("enter the distance to travel : "))
age = int(input("enter passanger's age : "))
fare = distance*5
if(distance>100 and (age<12 or age>60)):
    fare = fare - (fare*0.5)
    print("the ticket fare after discount is : ",fare)
elif(distance<=100 and (age<12 or age>60)):
    fare = fare - (fare*0.25)
    print("the ticket fare after discount is : ",fare)
else :
    print("the ticket fare is : ",fare)
```
